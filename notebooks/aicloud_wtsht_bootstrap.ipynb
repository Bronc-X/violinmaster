## Seed Manifest Shape

Create `data/processed/manifests/seed.jsonl` once you upload licensed audio and score files. Keep real audio and licensed score files out of Git.

```python
import json
from pathlib import Path

manifest_path = Path("data/processed/manifests/seed.example.jsonl")
example = {
    "id": "example-reference-001",
    "piece_id": "example-piece",
    "split": "train",
    "audio_path": "data/raw/audio/example-reference-001.wav",
    "musicxml_path": "data/raw/musicxml/example-piece.musicxml",
    "midi_path": "data/raw/midi/example-piece.mid",
    "performer_type": "reference",
    "rights_status": "licensed",
    "labels": {
        "pitch_quality": "stable",
        "rhythm_quality": "stable",
    },
}
manifest_path.write_text(json.dumps(example, ensure_ascii=False) + "\n", encoding="utf-8")
print(manifest_path.read_text(encoding="utf-8"))
```
